## Load data

```python
import pandas as pd
# comments
# let's do a sample first
comment_data_file = '../../data/nyt_comments/CommentsApril2017.csv'
article_data_file = '../../data/nyt_comments/ArticlesApril2017.csv'
comment_data = pd.read_csv(comment_data_file, sep=',')
article_data = pd.read_csv(article_data_file, sep=',')
print(comment_data.columns)
print(article_data.columns)
display(comment_data.head())
```

How many comments per topic?

```python
display(comment_data.loc[:, 'sectionName'].value_counts())
```

Unsurprising: most of the comments are directed toward politics and opinion pieces (`Sunday Review`).

Let's identify all the questions in the comments first.

```python
## clean text
import re
def clean_txt_matchers(text, txt_matcher_pairs):
    clean_text = text
    for txt_matcher, txt_sub in txt_matcher_pairs:
        clean_text = txt_matcher.sub(txt_sub, clean_text)
    return clean_text
HTML_matcher = re.compile('<[a-zA-Z0-9/]+>')
txt_matcher_pairs = [(HTML_matcher, ' ')]
comment_txt_var = 'commentBody'
comment_data = comment_data.assign(**{
    comment_txt_var : comment_data.loc[:, comment_txt_var].apply(lambda x: clean_txt_matchers(x, txt_matcher_pairs))
})
```

```python
# tokenize sentences
from nltk.tokenize import PunktSentenceTokenizer, WordPunctTokenizer
import re
sent_tokenizer = PunktSentenceTokenizer()
question_matcher = re.compile('\?$')
# testing
# test_sent = sent_tokenizer.tokenize('this is a sentence. is this another one?')
# print(list(filter(lambda x: question_matcher.search(x) is not None, test_sent)))
comment_data = comment_data.assign(**{
    'comment_questions' : comment_data.loc[:, comment_txt_var].apply(lambda x: list(filter(lambda y: question_matcher.search(y) is not None, sent_tokenizer.tokenize(x))))
})
# limit to questions with minimum token count
word_tokenizer = WordPunctTokenizer()
min_question_len = 5
comment_data = comment_data.assign(**{
    'comment_questions' : comment_data.loc[:, 'comment_questions'].apply(lambda x: list(filter(lambda y: len(word_tokenizer.tokenize(y)) >= min_question_len, x)))
})
```

```python
comments_with_question_data = comment_data[comment_data.loc[:, 'comment_questions'].apply(lambda x: len(x) > 0)]
print(comments_with_question_data.loc[:, 'comment_questions'].iloc[:10].values)
```

```python
## combine data
comment_article_data = pd.merge(
    article_data.loc[:, ['articleID', 'headline', 'snippet']],
    comments_with_question_data.loc[:, ['articleID', 'comment_questions', 'commentBody', 'userDisplayName', 'userID', 'userLocation', 'sectionName']],
    on='articleID', how='right'
)
# drop duplicates??
comment_article_data.drop_duplicates(['articleID', 'commentBody'], inplace=True)
```

What kinds of words/phrases tend to characterize reporter responses?

```python
from nltk.tokenize import WordPunctTokenizer
from nltk.stem.wordnet import WordNetLemmatizer
from stop_words import get_stop_words
class LemmaTokenizer:
    def __init__(self):
        self.wnl = WordNetLemmatizer()
        self.tok = WordPunctTokenizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in self.tok.tokenize(doc)]
en_stops = get_stop_words('en')
# tokenizer = WordPunctTokenizer()
# lemmatizer = WordNetLemmatizer()
tokenizer = LemmaTokenizer()
comment_data_text = comment_data.loc[:, 'commentBody'].values.tolist()
# comment_data_text = comment_data_text.apply(lambda x: ' '.join(list(map(lambda y: lemmatizer.lemmatize(y), tokenizer.tokenize(x)))))
# convert to DTM
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(min_df=0.00001, max_df=0.75, stop_words=en_stops, tokenizer=tokenizer)
comment_text_dtm = cv.fit_transform(comment_data_text)
```

```python
sorted_vocab = sorted(cv.vocabulary_.keys(), key=cv.vocabulary_.get)
comment_text_word_counts = pd.Series(np.array(comment_text_dtm.todense().sum(axis=0))[0, :], index=sorted_vocab).sort_values(inplace=False, ascending=False)
comment_text_word_pct = comment_text_word_counts / comment_text_word_counts.sum()
display(comment_text_word_pct.head(10))
```

```python
import numpy as np
reporter_comment_text_dtm = comment_text_dtm[np.where(comment_data.loc[:, 'commentType']=='reporterReply')[0], :]
print(reporter_comment_text_dtm.shape)
reporter_comment_text_word_counts = pd.Series(np.array(reporter_comment_text_dtm.sum(axis=0))[0, :], index=sorted_vocab)
# avoid overfitting
min_word_count = 3
reporter_comment_text_word_counts = reporter_comment_text_word_counts[reporter_comment_text_word_counts >= min_word_count]
reporter_comment_text_word_pct = reporter_comment_text_word_counts / reporter_comment_text_word_counts.sum()
# compute ratio
reporter_comment_text_word_ratio = (reporter_comment_text_word_pct / comment_text_word_pct).sort_values(inplace=False, ascending=False)
print(reporter_comment_text_word_ratio.head(50))
```

OK! These counts are more related to topics but we do see a high density of information being provided (e.g. `organ`, `suicide`, `patient`, `technology`) which suggests that the reporters are at least trying to answer the questions and that this information could have been included.

### Determining "audience"
How do we define audience? We need to think of groups that would have consistent questions and would have a vested interest in receiving answers to the questions (e.g. `As an X, I need to know Y`).

- Demographics (age, gender, geography, occupation, political)
- Intent (information seeking, opinion sharing)
- Experience (level of prior knowledge/involvement about the discussion)

Let's see if we can get at some of these factors by limiting our data to "I" sentences.

```python
first_person_pronouns = set(['I', 'me', 'my', "I'm"])
# tokenize
from nltk.tokenize import WordPunctTokenizer
tokenizer = WordPunctTokenizer()
comment_article_data = comment_article_data.assign(**{
    'comment_tokens' : comment_article_data.loc[:, 'commentBody'].apply(lambda x: tokenizer.tokenize(x))
})
first_person_comment_data = comment_article_data[comment_article_data.loc[:, 'comment_tokens'].apply(lambda x: len(set(x) & set(first_person_pronouns)) > 0)]
print('%d/%d comments with first-person mentions'%(first_person_comment_data.shape[0], comment_data.shape[0]))
```

### Testing demographics for users: gender
How many user names have associated demographic information? This could help us understand consistent differences among reader questions.

```python
print(comment_data.loc[:, 'userDisplayName'].value_counts(normalize=True).head(20))
```

Let's identify gender as a demographic based on common male/female names.

```python
import re
import os
name_data_dir = '../../data/meta_data/baby_names/'
# let's get all names from likely periods of birth for comments, i.e. 1930-2000
name_data_matcher = re.compile('yob19[3-9][0-9]')
name_data_files = list(filter(lambda x: name_data_matcher.search(x), os.listdir(name_data_dir)))
name_data_files = list(map(lambda x: os.path.join(name_data_dir, x), name_data_files))
name_data = pd.concat(list(map(lambda x: pd.read_csv(x, sep=',', header=None, index_col=False), name_data_files)), axis=0)
name_data.columns = ['name', 'gender', 'count']
# group by name, get raw count
name_count_data = name_data.groupby(['name', 'gender']).apply(lambda x: x.loc[:, 'count'].sum()).reset_index().rename(columns={0:'count'})
# # get gender percent
name_gender_data = name_count_data.groupby('name').apply(lambda x: x.assign(**{'count_pct' : x.loc[:, 'count'] / x.loc[:, 'count'].sum()}).drop(['count',], axis=1)).reset_index(drop=True)
def assign_label_by_cutoff_pct(data, label_var='gender', min_pct=0.75):
    data.sort_values('count_pct', inplace=True, ascending=False)
    cutoff_data = data[data.loc[:, 'count_pct'] >= min_pct]
#     print(data)
#     if(cutoff_data.loc[:, 'name'].iloc[0]=='Susan'):
#         print(cutoff_data)
    label = 'UNK'
    if(cutoff_data.shape[0] > 0):
        label = cutoff_data.iloc[0, :].loc[label_var]
    return label
min_gender_pct = 0.75
name_gender_label_data = []
for name_i, data_i in name_gender_data.groupby('name'):
    label_i = assign_label_by_cutoff_pct(data_i, label_var='gender', min_pct=min_gender_pct)
    name_gender_label_data.append([name_i, label_i])
name_gender_label_data = pd.DataFrame(name_gender_label_data, columns=['name', 'gender'])
# lowercase for consistency
name_gender_label_data = name_gender_label_data.assign(**{'name' : name_gender_label_data.loc[:, 'name'].apply(lambda x: x.lower())})
```

```python
## combine with comment data
# extract first names from comment user names
import re
def extract_name(text, camel_matcher):
    name = text
    text_tokens = text.split(' ')
    if(len(text_tokens) > 0):
        name = text_tokens[0]
    elif(camel_matcher.search(text) is not None):
        name = camel_matcher.search(text).group(0)
    name = name.lower()
    return name
camel_matcher = re.compile('[a-z]+(?=[A-Z])')
comment_article_data = comment_article_data.assign(**{
    'user_first_name' : comment_data.loc[:, 'userDisplayName'].apply(lambda x: extract_name(str(x), camel_matcher))
})
print(comment_article_data.loc[:, 'user_first_name'].value_counts())
name_gender_label_data.rename(columns={'name' : 'user_first_name'}, inplace=True)
# add gender lol
if('gender' in comment_article_data.columns):
    comment_article_data.drop('gender', axis=1, inplace=True)
comment_article_data = pd.merge(comment_article_data, name_gender_label_data, on='user_first_name', how='left')
comment_article_data.fillna(value={'gender' : 'UNK'}, inplace=True)
display(comment_article_data.loc[:, ['user_first_name', 'gender']])
# get top names per gender
for gender_i, data_i in comment_article_data.groupby('gender'):
    print(f'top names for gender={gender_i}')
    print(data_i.loc[:, 'user_first_name'].value_counts().iloc[:10])
```

### Testing demographics for users: geography
Do people in different locations tend to ask different questions?

Start simple: compare US vs. non-US.

```python
print(comment_article_data.loc[:, 'userLocation'].value_counts().head(20))
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
geo_locator = Nominatim(user_agent='test_app_name')
geo_coder = RateLimiter(geo_locator.geocode, min_delay_seconds=0.5)
filter_locations = set(['<br/>'])
user_locations = comment_article_data[~comment_article_data.loc[:, 'userLocation'].isin(filter_locations)].loc[:, 'userLocation'].drop_duplicates()
print('%d unique locations to parse'%(len(user_locations)))
from tqdm import tqdm
tqdm.pandas()
user_geo_code_locations = user_locations.progress_apply(geo_coder)
```

```python
# let's see the distribution of countries
import re
def assign_country(location, matcher_code_pairs):
    country = 'UNK'
    if(location is not None):
        country = 'other'
        location_address = location.address
        for matcher, code in matcher_code_pairs:
            if(matcher.search(location_address) is not None):
                country = code
                break
    return country
us_country_matcher = re.compile('United States of America$')
us_country_code = 'US'
matcher_code_pairs = [(us_country_matcher, us_country_code)]
user_geo_code_location_country = user_geo_code_locations.apply(lambda x: assign_country(x, matcher_code_pairs))
print(user_geo_code_location_country.value_counts())
# add to original data
user_location_data = pd.concat([user_locations, user_geo_code_location_country], axis=1)
user_location_data.columns = ['userLocation', 'user_country']
if('user_country' not in comment_article_data.columns):
    comment_article_data = pd.merge(comment_article_data, user_location_data, )
display(user_location_data.head())
```

Out of the users who provided their location...let's see if we can differentiate questions posed by US readers and non-US readers.

### Testing experience among users
Generating questions for writers could depend on the audience's expected level of experience with the issue.

Do more-experienced users tend to ask more critical questions? Do these users tend toward more political or personal articles? 

We need more data for this! Let's load 1 year of comment data and identify more/less experienced commenters based on their activity.

```python
import pandas as pd
import re
import os
from datetime import datetime
comment_data_file_matcher = re.compile('Comments.*csv')
data_dir = '../../data/nyt_comments/'
comment_data_files = list(filter(lambda x: comment_data_file_matcher.search(x) is not None, os.listdir(data_dir)))
comment_data_files = list(map(lambda x: os.path.join(data_dir, x), comment_data_files))
# print(comment_data_files)
# load, combine comment data
month_matcher = re.compile('[A-Z][a-z]+(?=[0-9])')
year_matcher = re.compile('[0-9]{4}')
months = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_lookup = dict(zip(months, range(1,len(months)+1)))
combined_comment_data = []
for comment_data_file_i in comment_data_files:
    comment_data_i = pd.read_csv(comment_data_file_i, sep=',', index_col=False, usecols=['approveDate', 'commentBody', 'commentID', 'commentType', 'userDisplayName', 'userID', 'sectionName', 'articleID'])
    comment_data_date_i = [month_lookup[month_matcher.search(comment_data_file_i).group(0)], int(year_matcher.search(comment_data_file_i).group(0))]
    comment_data_date_i = datetime(day=1, month=comment_data_date_i[0], year=comment_data_date_i[1])
    comment_data_i = comment_data_i.assign(**{'comment_time_period' : comment_data_date_i})
    combined_comment_data.append(comment_data_i)
combined_comment_data = pd.concat(combined_comment_data, axis=0)
# add article data lol
article_data_file_matcher = re.compile('Articles.*csv')
article_data_files = list(filter(lambda x: article_data_file_matcher.search(x) is not None, os.listdir(data_dir)))
article_data_files = list(map(lambda x: os.path.join(data_dir, x), article_data_files))
article_data = pd.concat(list(map(lambda x: pd.read_csv(x, sep=',', index_col=False, usecols=['articleID', 'headline', 'snippet', 'pubDate']), article_data_files)))
combined_comment_data = pd.merge(combined_comment_data, article_data, on='articleID')
# fix data types
# article_date_format = '2018-04-21 18:30:10'
article_date_format = '%Y-%m-%d %H:%M:%S'
combined_comment_data = combined_comment_data.assign(**{
    'userID' : combined_comment_data.loc[:, 'userID'].astype(int),
    'commentID' : combined_comment_data.loc[:, 'commentID'].astype(int),
    'approveDate': combined_comment_data.loc[:, 'approveDate'].apply(lambda x: datetime.fromtimestamp(x)),
    'pubDate': combined_comment_data.loc[:, 'pubDate'].apply(lambda x: datetime.strptime(x, article_date_format)),
})
print(combined_comment_data.shape[0])
display(combined_comment_data.head())
```

```python
from datetime import datetime
prior_comment_date_cutoff = datetime(day=1, month=3, year=2018)
user_id_var = 'userID'
prior_comment_counts = combined_comment_data[combined_comment_data.loc[:, 'comment_time_period'] <= prior_comment_date_cutoff].loc[:, user_id_var].value_counts()
prior_comment_counts = np.log(prior_comment_counts)
min_pct_cutoff = 50
max_pct_cutoff = 95
min_comment_cutoff = np.percentile(prior_comment_counts, min_pct_cutoff)
max_comment_cutoff = np.percentile(prior_comment_counts, max_pct_cutoff)
print('min comment cutoff = %.3f'%(np.exp(min_comment_cutoff)))
print('max comment cutoff = %.3f'%(np.exp(max_comment_cutoff)))
prior_commenters = set(prior_comment_counts.index)
non_active_commenters = set(prior_comment_counts[prior_comment_counts <= min_comment_cutoff].index.tolist())
active_commenters = set(prior_comment_counts[prior_comment_counts >= max_comment_cutoff].index.tolist())
print('%d active commenters'%(len(active_commenters)))
```

```python
# limit to recent data
recent_comment_data = combined_comment_data[combined_comment_data.loc[:, 'comment_time_period'] > prior_comment_date_cutoff]
top_k = 10
article_topic_counts = combined_comment_data.loc[:, 'sectionName'].value_counts()
top_k_topics = article_topic_counts.index.tolist()[:top_k]
sample_articles_per_topic = 10
sample_article_data = []
np.random.seed(123)
for topic_i in top_k_topics:
    print(f'*~*~*~* topic={topic_i} *~*~*~*')
    data_i = recent_comment_data[recent_comment_data.loc[:, 'sectionName']==topic_i]
    # make sure that each article has at least one non-active and one active commenter
    sample_ctr_i = 0
    np.random.shuffle(data_i.values)
    article_ids_i = data_i.loc[:, 'articleID'].unique()
    article_ctr = 0
    for article_id_j in article_ids_i:
        data_j = data_i[data_i.loc[:, 'articleID']==article_id_j]
        article_date_j = data_j.loc[:, 'pubDate'].iloc[0]
        headline_j = data_j.loc[:, 'headline'].iloc[0]
        # check for active commenter
        active_commenters_j = list(set(data_j.loc[:, 'userID'].unique()) & active_commenters)
        active_commenter_data_j = data_j[data_j.loc[:, 'userID'].isin(active_commenters)]
        if(len(active_commenters_j) > 0 and data_j.shape[0] > 1):
            # check for non-active commenter
            prior_data_j = data_i[data_i.loc[:, 'approveDate'] <= article_date_j]
            prior_commenters_j = set(prior_data_j.loc[:, 'userID'].unique())
            non_active_commenters_j = list(set(data_j.loc[:, 'userID'].unique()) - active_commenters)
            for non_active_commenter_l in non_active_commenters_j:
                has_prior_activity_l = (non_active_commenter_l in active_commenters or non_active_commenter_l in prior_commenters_j)
                if(not has_prior_activity_l):
                    non_active_commenter_data_j = data_j[data_j.loc[:, 'userID']==non_active_commenter_l]
                    sample_non_active_commenter_data_j = non_active_commenter_data_j.iloc[0, :]
                    sample_active_commenter_data_j = active_commenter_data_j.iloc[0, :]
                    print('+++ headline=%s +++'%(headline_j))
                    print('\tactive commenter=%s\n\tcomment=%s'%
                          (sample_active_commenter_data_j.loc['userDisplayName'], 
                           sample_active_commenter_data_j.loc['commentBody']))
                    print('\tnon-active commenter=%s\n\tcomment=%s'%
                          (sample_non_active_commenter_data_j.loc['userDisplayName'], 
                           sample_non_active_commenter_data_j.loc['commentBody']))
                    break
        article_ctr += 1
        if(article_ctr > sample_articles_per_topic):
            break
```

- `Unknown`: active commenters seem to engage more with article content (?)
- `Politics`: active commenters have more complicated/constructive comments, less (?) self-disclosure
- `Sunday Review`: active commenters have more background-information comments, similar amounts of self-disclosure
- `Europe`: active commenters have more (?) rhetorical questions; all commenters engage with article content
- `Middle East`: all commenters have similar amount of criticism toward authors
- `Asia Pacific`: active commenters have more background-information comments
- `Media`: active commenters have more complicated comments
- `Family`: active commenters have less (?) self-disclosure, less complicated comments
- `Television`: active commenters have similar amount of background-information comments

In general, it's hard to see consistent differences between the groups, although at a high level we may end up seeing e.g. more critical behavior overall (negative sentiment, addressing concepts that weren't in the article).

### Testing user values
Some users may have differences in underlying values, e.g. issues related to family or to community as opposed to politics. How readily do people express these values in comments, and would it help the writer to know about these values?

These "values" are going to be harder to annotate. Can we get a first pass using LIWC data?

```python
import pandas as pd
LIWC_data = pd.read_csv('/home/public/LIWC_analysis/LASL-v3/res/LIWC.all.txt', sep=',', index_col=False, header=None)
LIWC_data.columns = ['LIWC_pattern', 'LIWC_class']
LIWC_data = LIWC_data.assign(**{
    'LIWC_pattern' : LIWC_data.loc[:, 'LIWC_pattern'].apply(lambda x: x.replace('*', '.*').strip())
})
print(LIWC_data.head())
print(LIWC_data.loc[:, 'LIWC_class'].value_counts().head(50))
```

```python
value_categories = ['AFFECT', 'SOCIAL', 'COGMECH', 'OCCUP', 'INSIGHT', 'JOB', 
                    'RELIG', 'FAMILY', 'HUMANS', 'SWEAR', 'NEGATE', 'FRIENDS']
LIWC_matchers = LIWC_data.groupby('LIWC_class').apply(lambda x: re.compile('^(%s)$'%('|'.join(x.loc[:, 'LIWC_pattern']))))
# tokenize comment data
from nltk.tokenize import WordPunctTokenizer
tokenizer = WordPunctTokenizer()
comment_article_data = comment_article_data.assign(**{
    'comment_tokens' : comment_article_data.loc[:, 'commentBody'].apply(lambda x: tokenizer.tokenize(x.lower()))
})
# match LIWC tokens
from pandarallel import pandarallel
JOBS=10
pandarallel.initialize(nb_workers=JOBS)
comment_LIWC_data = comment_article_data.loc[:, ['articleID', 'headline', 'commentBody', 'commentID', 'userID', 'comment_tokens', 'sectionName', 'comment_questions']]
for value_category_i in value_categories:
    category_matcher_i = LIWC_matchers.loc[value_category_i]
    comment_LIWC_data = comment_LIWC_data.assign(**{
        f'{value_category_i}_tokens' : comment_LIWC_data.loc[:, 'comment_tokens'].parallel_apply(lambda x: list(filter(lambda y: category_matcher_i.match(y), x)))
    })
    comment_LIWC_data = comment_LIWC_data.assign(**{
        f'{value_category_i}' : comment_LIWC_data.apply(lambda x: len(x.loc[f'{value_category_i}_tokens']) / len(x.loc['comment_tokens']), axis=1)
    })
# flatten data for easier handling later ;_;
comment_LIWC_data = comment_LIWC_data.drop_duplicates(['articleID', 'commentBody'], inplace=False)
comment_LIWC_data = comment_LIWC_data.assign(**{'commentID' : list(range((comment_LIWC_data.shape[0])))})
# value_category_pct_vars = [f'{c}_pct' for c in value_categories]
flat_comment_LIWC_data = comment_LIWC_data.loc[:, ['articleID', 'commentID','sectionName'] + value_categories]
flat_comment_LIWC_data = pd.melt(flat_comment_LIWC_data, value_vars=value_categories, id_vars=['articleID',  'commentID','sectionName'], var_name='word_category', value_name='word_pct')
display(flat_comment_LIWC_data.head())
```

```python
# let's look at the distribution of LIWC percentages
# collect mean LIWC percent per topic
top_k = 10
topic_counts = comment_article_data.loc[:, 'sectionName'].value_counts()
top_k_topics = topic_counts.index.tolist()[:top_k]
top_k_topic_comment_LIWC_data = flat_comment_LIWC_data[flat_comment_LIWC_data.loc[:, 'sectionName'].isin(top_k_topics)]
top_k_topic_comment_LIWC_pct = top_k_topic_comment_LIWC_data.groupby(['sectionName', 'word_category']).apply(lambda x: x.loc[:, 'word_pct'].mean()).reset_index(name='word_pct')
# un-flatten
top_k_topic_comment_LIWC_pct = pd.pivot(top_k_topic_comment_LIWC_pct, index='sectionName', columns='word_category', values='word_pct')
display(top_k_topic_comment_LIWC_pct.head())
# plot
import seaborn as sns
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
sns.heatmap(data=top_k_topic_comment_LIWC_pct, cmap='Blues')
plt.show()
```
